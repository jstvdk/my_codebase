```python
def surface_brightness(energy_start, energy_end):
    result = []
    for i in range(0,10):
        for j in range(0,10):
            test = pd.read_csv(f"/Users/vdk/TuCluster/galprop_runs/tab_TeV_halos_2/tab/dphi_theta{i}.{j}.csv")
            dphi_table = test.to_numpy()
            sum = 0
            dphi_table_work = dphi_table[energy_start:energy_end]
            for b,k in enumerate(dphi_table_work):
                if b == len(dphi_table_work)-1:
                    dE = dphi_table_work[b][0]-dphi_table_work[b-1][0]
                    sum += k[0]*k[1]*2*math.pi*math.sin(math.radians(i+(j+1)/10))*dE
                else:
                    dE= dphi_table_work[b+1][0]-dphi_table_work[b][0]
                    sum += k[0]*k[1]*2*math.pi*math.sin(math.radians(i+(j+1)/10))*dE
            result.append((sum/(math.pi*(i+(j+1)/10)**2))*10**6) # ділимо на площу pi*deg^2 і множимо на 10^6 що б перевести в еВ?
    return result

x = [i for i in range(1,101)]


SB = surface_brightness(0,50) # початок включно, кінець не включно

plt.figure(figsize=(10,10))
plt.loglog(x, SB)
```
