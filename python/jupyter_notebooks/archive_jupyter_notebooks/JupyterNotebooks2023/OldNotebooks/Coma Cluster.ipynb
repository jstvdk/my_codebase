```python
energy = np.logspace(-2,6,100)*u.GeV
e1, s1 = clust_Coma_CRpProptoN.get_gamma_spectrum(energy,Rmax=clust_Coma_CRpProptoN.R500, type_integral='spherical')
e2, s2 = clust_Coma_CRpProptoN05.get_gamma_spectrum(energy,Rmax=clust_Coma_CRpProptoN05.R500, type_integral='spherical')
e3, s3 = clust_Coma_alpha28.get_gamma_spectrum(energy,Rmax=clust_Coma_alpha28.R500, type_integral='spherical')
e4, s4 = clust_Coma_alpha33.get_gamma_spectrum(energy,Rmax=clust_Coma_alpha33.R500, type_integral='spherical')
e5, s5 = clust_Coma_alpha28_N05.get_gamma_spectrum(energy,Rmax=clust_Coma_alpha33.R500, type_integral='spherical')
e6, s6 = clust_Coma_alpha33_N05.get_gamma_spectrum(energy,Rmax=clust_Coma_alpha33.R500, type_integral='spherical')





plt.gca().set_ylim(bottom= 1e-15)
plt.gca().set_ylim(top= 1e-9)
plt.gca().set_xlim(left= 1e-2)
plt.gca().set_xlim(right= 1e6)

plt.loglog(e1, (e1**2*s1).to('GeV cm-2 s-1'), '-.', label='$n_{CRp} \propto n_e$', c = 'r')
plt.loglog(e2, (e2**2*s2).to('GeV cm-2 s-1'), label= r' $ n_{CRp} \propto n^{1/2}_e , \alpha = -2.3 $', c = 'r')

plt.loglog(e3, (e3**2*s3).to('GeV cm-2 s-1'), '-.', c = 'b')
plt.loglog(e3, (e5**2*s5).to('GeV cm-2 s-1'), label=r' $ n_{CRp} \propto n^{1/2}_e , \alpha = -2.8 $', c = 'b')

plt.loglog(e4, (e4**2*s4).to('GeV cm-2 s-1'), '-.', c = 'g')
plt.loglog(e4, (e6**2*s6).to('GeV cm-2 s-1'), label=r' $ n_{CRp} \propto n^{1/2}_e , \alpha = -3.3 $', c = 'g')
plt.xlabel('Energy (GeV)')
plt.ylabel('Flux $(GeV \: cm^{-2} \: s^{-1}$)')
plt.legend()
```
