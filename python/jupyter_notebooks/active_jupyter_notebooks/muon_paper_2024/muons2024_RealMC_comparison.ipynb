```python
import numpy as np
from scipy import stats

def match_distributions(list1, list2, num_elements):
    # Calculate the histograms and CDFs
    hist1, bin_edges1 = np.histogram(list1, bins='auto', density=True)
    cdf1 = np.cumsum(hist1)
    hist2, bin_edges2 = np.histogram(list2, bins='auto', density=True)
    cdf2 = np.cumsum(hist2)
    
    # Normalize the CDFs
    cdf1 = cdf1 / cdf1[-1]
    cdf2 = cdf2 / cdf2[-1]

    # Use the inverse CDF (quantile function) to map values
    quantiles1 = np.interp(np.linspace(0, 1, num_elements), cdf1, bin_edges1[:-1])
    quantiles2 = np.interp(np.linspace(0, 1, num_elements), cdf2, bin_edges2[:-1])
    
    # Generate new lists with matched distributions
    new_list1 = np.interp(list1, bin_edges1[:-1], quantiles1)
    new_list2 = np.interp(list2, bin_edges2[:-1], quantiles2)
    
    return new_list1, new_list2

# Example usage
list1 = np.random.normal(loc=10, scale=2, size=1000)  # More values at higher numbers
list2 = np.random.normal(loc=5, scale=2, size=1000)   # More values at lower numbers

new_list1, new_list2 = match_distributions(list1, list2, min(len(list1), len(list2)))

# Verify the distributions
import matplotlib.pyplot as plt
plt.hist(new_list1, bins=30, alpha=0.5, label='New List 1')
plt.hist(new_list2, bins=30, alpha=0.5, label='New List 2')
plt.legend()
plt.show()
```
